## ANOVA 
#### Analysis Of Variance
#### F-Test with P-Value

```python
import pandas as pd
```

```python
dataset = pd.read_csv("PrePlacement.csv")
```

```python
dataset
```

### Statement 
Is there any significant difference between ssc_p, hsc_p & degree_p ? 

### State of hypothesis 

Ho = There is no significant differnce betwween ssc_p,hsc_p,degree_p

H1 = There is significant differnce betwween ssc_p,hsc_p,degree_p

### Significance Level
P < 0.05

### ANOVA --> One-way Classification (Either Row-wise / column-wise) 

```python
from scipy.stats import f_oneway
f_oneway(dataset['ssc_p'],dataset['hsc_p'],dataset['degree_p'])
```

### Inference 
pvalue = 0.49

pvalue > 0.05, so we Accept the Null hypothesis

==> There is no significant differences between ssc_p, hsc_p & degree_p


### ANOVA --> Two-way Classification (Takes both Row & columns) 

In Scipy, there is no direct method for two-way ANOVA (as we have for one-way ANOVA using f_oneway).

### ANOVA --> Two-Way Classification with statsmodels

```python
import statsmodels.api as sm              # This line imports the main interface of the statsmodels library and assigns it the alias sm
from statsmodels.formula.api import ols     # This line specifically imports the ols function from the formula.api module of statsmodels

# The formula is 'dependent_variable ~ independent_variable'
model = ols('sl_no ~ ssc_p + hsc_p + etest_p + ssc_p * hsc_p * etest_p', data = dataset).fit() 
ANOVA_Table = sm.stats.anova_lm(model)
```

```python
ANOVA_Table
```

### Inference 

P-value = 0.51

P-valiue > 0.05 

Accept the Null hypothesis

Since it is two-way classification --> There is no significant difference between ssc_p, hsc_p and etest_p  percentages of all the students.

Row = sl_no  
Columns = ssc_p,  hsc_p,  etest_p
