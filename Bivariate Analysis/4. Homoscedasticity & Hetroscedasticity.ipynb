## Homoscedasticity & Hetroscedasticity 
#### By finding & ploting the Residuals 

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Salary_Data.csv")
```

```python
data
```

```python
dataset = pd.DataFrame(data)
```

```python
dataset
```

```python
indep = dataset[['YearsExperience']]
dep = dataset[['Salary']]
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(indep,dep) 
```

```python
weight = regressor.coef_
weight
```

```python
bias = regressor.intercept_
bias
```

```python
Predicted_values = regressor.predict(indep)
```

### Check for the Residuals 
Error / Residual = Actual values - Predicted values

```python
Residual = indep - Predicted_values
Residual
```

### Plot the graph for Residuals to identify whether it is Homoscedasticity / Heteroscedasticity

```python
plt.scatter(Predicted_values,Residual)
plt.axhline(y=0, color='r', linestyle='--')   # Red Line: Represents the zero line for residuals
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.title('Residuals vs. Predicted Values')
plt.show()
```

### Inference for the above graph : There is no pattern in this graph, so it is Homoscedasticity

### Thumb Rule - Homoscedasticity (Same Variance)
when we plot all the residual values against the Predicted values (Fitted values / independent variables), in repetition it follows a constant spread (variance) for any kind of Intervals.


### Thumb Rule - Hetroscedasticity (Different Variance)
when we plot all the residual values against the Predicted values (Fitted values / independent variables), if the spread increases or decreases (funel shape), Clustering of points, bow shaped or without any pattern, it is considered as Hetroscedasticity.

### When to check for Homoscedasticity
1. Regression Analysis - Before finalizing a model it is one of the key assumptions, to have a good practice to check for homoscedasticity to identify whether the variance of the error is constant without any patterns.

2. Model Evaluation - can be more confident in validating the statistical test(T-test)


### When to check for Homoscedasticity
1. Unexpected Patterns in Residuals - it indicates the presence of heteroscedasticity, This means we have to go for further investigation.

### Which is Good among these two
Homoscedasticity is Preferable : For regression models because it aligns with the assumptions of Machine Learning models and ensures reliable statistical inference.
