```python
quan = []
qual = []
for columnName in dataset.columns:
    if dataset[columnName].dtype == 'object':
        qual.append(columnName)
    else:
        quan.append(columnName)
```

```python
quan
```

```python
qual
```

```python
for columnName in quan:
    print(columnName)
```

```python
for columnName in quan:
    dataset[columnName].fillna(dataset[columnName].mean(),inplace=True)
```

```python
dataset
```

```python
dataset.isnull().sum()
```

```python
for columnName in qual:
    dataset[columnName].fillna(dataset[columnName].mode()[0],inplace=True)
```

```python
dataset
```

```python
dataset.isnull().sum()
```

```python
import numpy as np
from sklearn.impute import SimpleImputer
quan_imputer = SimpleImputer(missing_values = np.nan, strategy = 'mean')
qual_imputer = SimpleImputer(missing_values = np.nan, strategy = 'most_frequent')

# Combine the imputers using ColumnTransformer
# List of (name, transformer, columns) tuples specifying the transformer objects to be applied to subsets of the data.
from sklearn.compose import ColumnTransformer
Preprocessor = ColumnTransformer(transformers = [['quan',quan_imputer,quan],['qual',qual_imputer,qual]])

# Fit the imputers and transform the dataset
dataset_imputed = Preprocessor.fit_transform(dataset)

# Convert the result back to a DataFrame
dataset_NOnull = pd.DataFrame(dataset_imputed, columns = quan + qual)

dataset_NOnull
```

```python
dataset_NOnull.isnull().sum()
```

```python
dataset_NOnull.dtypes
```

```python
# Drawback: When you use ColumnTransformer combined with SimpleImputer, the output is a NumPy array,
#            and all data is cast to a general object type if you have both numerical and categorical columns.

# To fix this issue and maintain the correct data types for each column, you can manually convert the numerical and categorical columns 
# back to their appropriate types after the imputation process.

dataset_NOnull[quan] = dataset_NOnull[quan].apply(pd.to_numeric)
```
