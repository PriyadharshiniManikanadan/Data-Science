```python
quan,qual
```

```python
# Logic into the Function

def quanQual(dataset):
    qual = []
    quan = []
    for columnName in dataset.columns:
        #print(columnName)
        if dataset[columnName].dtype == 'object':
            #print("qual")
            qual.append(columnName)
        else:
            #print("quan")
            quan.append(columnName)
    return quan,qual
```

```python
# check the Function
quan,qual = quanQual(dataset)
quan,qual
```

```python
# Check Mean 
Descriptive = pd.DataFrame(index=['Mean','Median','Mode'], columns = quan)
for columnName in quan:
    Descriptive[columnName]['Mean'] = dataset[columnName].mean()
Descriptive
```

```python
# Check for Median & Mode 
Descriptive = pd.DataFrame(index=['Mean','Median','Mode'], columns = quan)
for columnName in quan:
    Descriptive[columnName]['Mean'] = dataset[columnName].mean()    # First - Column , Second - Rows (Alphabetical order)
    Descriptive[columnName]['Median'] = dataset[columnName].median()
    Descriptive[columnName]['Mode'] = dataset[columnName].mode()[0]     # [0]  - to get only the numbers
Descriptive
```

## 3. Percentile

```python
import numpy as np
```

```python
np.percentile(dataset['ssc_p'],50)

# This output Value denotes that 50% of data exists within 67
# For this dataset we can say that 50% of the students obtained marks within 67 score
```

```python
Descriptive
```

```python
# Add Percentiles in Descriptive Table
Descriptive = pd.DataFrame(index=['Mean','Median','Mode','Q1:25%','Q2:50%',"Q3:75%","99%","Q4:100%"], columns = quan)
for columnName in quan:
    Descriptive[columnName]['Mean'] = dataset[columnName].mean()    # First - Column , Second - Rows (Alphabetical order)
    Descriptive[columnName]['Median'] = dataset[columnName].median()
    Descriptive[columnName]['Mode'] = dataset[columnName].mode()[0]     # [0]  - to get only the numbers
Descriptive
```

```python
# Check for all 4 Quadrants(Percentile) 

Descriptive = pd.DataFrame(index=['Mean','Median','Mode','Q1:25%','Q2:50%',"Q3:75%","99%","Q4:100%"], columns = quan)
for columnName in quan:
    Descriptive[columnName]['Mean'] = dataset[columnName].mean()    
    Descriptive[columnName]['Median'] = dataset[columnName].median()
    Descriptive[columnName]['Mode'] = dataset[columnName].mode()[0]     
    Descriptive[columnName]['Q1:25%'] = dataset.describe()[columnName]["25%"]
    Descriptive[columnName]['Q2:50%'] = dataset.describe()[columnName]["50%"]
    Descriptive[columnName]['Q3:75%'] = dataset.describe()[columnName]["75%"]
    Descriptive[columnName]['99%'] = np.percentile(dataset[columnName],99)
    Descriptive[columnName]['Q4:100%'] = dataset.describe()[columnName]["max"]
Descriptive
```
